```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# VOTES

# Read in voting data and set up initial voting dataframe

df_votes = pd.read_csv("../data_sources/elections_by_county_Eric.csv")
df_votes.head()
```

```python
# select oregon voting data
df_OR_votes = df_votes[df_votes.state == 'OR']
df_OR_votes.head()
```

```python
# Deaths

# read in Death data and prep initial dataframe

df_deaths = pd.read_csv("../data_sources/Provisional_COVID-19_Death_Counts_in_the_United_States_by_County.csv")

df_deaths= df_deaths.rename(columns={'Date as of':'Date-as-of', 'Start Date':'Start_date', 'End Date': 'End_date', 'State':'state',
                            'County name':'CountyName', 'FIPS County Code':'FIPS_code', 'Urban Rural Code':'Urban-Rural_code',
                           'Deaths involving COVID-19':'Deaths_COVID', 'Deaths from All Causes':'Deaths_All'})

df_deaths.head()
```

```python
#select Oregon Death statistics

df_OR_deaths = df_deaths[df_deaths.state == 'OR']
df_OR_deaths.head()
```

```python
df_OR_deaths[['county','Extra']] = df_OR_deaths.CountyName.str.split(" County",expand=True,)
df_OR_deaths.head()
```

```python
# Merge data frames
merged_orgeon = pd.merge(df_OR_votes,df_OR_deaths, how='outer', on='county')
```
